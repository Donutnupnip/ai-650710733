```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student.csv")
```

```python
df.head()
```

# ดูข้อมูลหลักที่จะใช้ทำ

```python
df
```

#ทำให้เพศเป็นมาตรฐานเดียวกัน

```python
gender_mapping = {
    'ช': 'ชาย',
    'ชาย': 'ชาย',
    'M': 'ชาย',
    'Male': 'ชาย',

    'ญ': 'หญิง',
    'หญิง': 'หญิง',
    'F': 'หญิง',
    'Female': 'หญิง'
}
df['เพศ'] = df['เพศ'].map(gender_mapping)
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
df
```

#แสดงข้อมูลก่อนเติมค่าเฉลี่ยอายุ




```python
print("ก่อนเติมค่า")
df
```

#แสดงข้อมูลหลังเติมค่าเฉลี่ยอายุ จะเห็นว่าค่าที่เคยเป็น NaN ตอนนี้เปลี่ยนให้กลายเป็นค่าเฉลี่ยแล้ว




```python
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)

df['อายุ'] = df['อายุ'].round(1)

print(f"\nค่าเฉลี่ยอายุ - {mean_age:.1f} ปี")
print("หลังเติมค่า")
df
```

#การจัดการค่าผิดปกติ


#ใช้ IQR (Interquartile Range)

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

#กำหนดขอบเขต

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

# กรองค่าผิดปกติ

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```

#เติมค่าด้วยค่าที่พบบ่อยที่สุด เราจะเติมในส่วนของ คณะ
#ก่อนเติม

```python
print("ก่อนเติมค่า")
df_no_outliers
```

#ใช้โค้ดเหมาะกับข้อมูลประเภท - ใช้ค่าที่พบบ่อยที่สุด (mode)
#หลังเติม จะสังเกตุว่าค่าที่เป็น NaN ถูกแทนค่าด้วยค่าที่พบมากที่สุด

```python
df_no_outliers['คณะ'].fillna(df_no_outliers['คณะ'].mode()[0], inplace=True)

print("\nหลังเติมค่า")
df_no_outliers
```

#การทำให้ข้อมูลเป็นมาตรฐาน ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'วิศวกรรม', 'วิศวฯ', 'วิศวกรรมศาสตร์', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

# ใช้งาน
df_no_outliers['คณะ'] = df_no_outliers['คณะ'].apply(standardize_faculty)
df_no_outliers
```

# แก้ไข GPA ผิดปกติ

```python
df_no_outliers.loc[df_no_outliers['GPA'] < 0, 'GPA'] = np.nan
df_no_outliers.loc[df_no_outliers['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df_no_outliers['GPA'].mean()
df_no_outliers['GPA'].fillna(mean_gpa, inplace=True)
df_no_outliers
```

# แก้ไขส่วนสูง น้ำหนัก

```python


df_no_outliers['น้ำหนัก_kg'].fillna(df_no_outliers['น้ำหนัก_kg'].median(), inplace=True)

df_no_outliers
```
